```python
import numpy as np
import pandas as pd
import random as rd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
training  = pd.read_csv('training.csv')
#training.columns
training = training.drop(labels =['Unnamed: 0','Customer ID'],axis =1)
training
```

```python
training['Streaming TV'] = training['Streaming TV'].fillna(method='bfill')
training['Streaming TV'] = training['Streaming TV'].fillna(method='pad')
```

```python
training.isnull().sum()
#training.to_csv('All_Feature_TrainIDdataset.csv',sep=',')
```

```python
x = training.iloc[:,:-1]
y = training.iloc[:,-1]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
clf = SVC(kernel='rbf',C=1,gamma='auto')
clf.fit(x_train,y_train)
y_predict = clf.predict(x_test)
print(clf.score(x_train,y_train))
print(clf.score(x_test,y_test))
```
