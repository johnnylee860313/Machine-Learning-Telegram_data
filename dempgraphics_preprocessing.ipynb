```python
import numpy as np
import pandas as pd
demographic = pd.read_csv('demographics.csv')
pd.set_option("display.max_rows", 20)
demographic
```

```python
demographic.isnull().sum()
```

```python
demographic['Gender']= demographic['Gender'].replace(['Male','Female'],[1,0])
demographic['Under 30']= demographic['Under 30'].replace(['Yes','No'],[1,0])
demographic['Senior Citizen']= demographic['Senior Citizen'].replace(['Yes','No'],[1,0])
demographic['Married']= demographic['Married'].replace(['Yes','No'],[1,0])
demographic['Dependents']= demographic['Dependents'].replace(['Yes','No'],[1,0])

demographic
```

```python
Under30_avg = demographic[demographic['Age']<=30].mean()
Under30_avg = Under30_avg['Age']
Senior_Citizen = demographic[demographic['Age']>=65].mean()
Senior_Citizen = Senior_Citizen['Age']
print(Under30_avg)
print(Senior_Citizen)
```

```python
tempt = demographic.loc[demographic['Married']==1]
tempt = tempt.loc[tempt['Age']>30]
tempt = tempt.loc[tempt['Age']<65]
tempt
```

```python
Married_between30and65 = tempt['Age'].mean()
Married_between30and65
```

```python
tempt = demographic.loc[demographic['Married']==0]
tempt = tempt.loc[tempt['Age']>30]
tempt = tempt.loc[tempt['Age']<65]
tempt
```

```python
Unmarried_between30and65 = tempt['Age'].mean()
Unmarried_between30and65
```

```python
Age_mean = demographic['Age'].mean()
Age_mean
```

```python
demographic = demographic.drop(labels=['Count'],axis =1)
demographic
```

```python
demographic['Gender'] = demographic['Gender'].fillna(demographic['Gender'].mode()[0])
demographic
```

```python
for idx,cell in demographic['Age'].iteritems():
    if np.isnan(cell):
        under30 = demographic['Under 30'][idx]
        sc = demographic['Senior Citizen'][idx]
        m = demographic['Married'][idx]
        if(not(np.isnan(under30))):
            if(under30):
                demographic['Age'][idx]=Under30_avg
            elif(not(np.isnan(sc))):
                if(sc):
                    demographic['Age'][idx]=Senior_Citizen
                elif(not(np.isnan(m))):
                    if(m):
                        demographic['Age'][idx]=Married_between30and65
                    else:
                        demographic['Age'][idx]=Unmarried_between30and65
        else :
            demographic['Age'][idx]=Age_mean

demographic['Age'] = demographic['Age'].fillna(demographic['Age'].mean())
demographic['Age'].isnull().sum()
```

```python
for idx,cell in demographic['Under 30'].iteritems():
    if(np.isnan(cell)):
        if((demographic['Age'][idx])>30):
            demographic['Under 30'][idx]=0
        else:
            demographic['Under 30'][idx]=1

for idx,cell in demographic['Senior Citizen'].iteritems():
    if(np.isnan(cell)):
        if((demographic['Age'][idx])>65):
            demographic['Senior Citizen'][idx]=1
        else:
            demographic['Senior Citizen'][idx]=0

demographic.isnull().sum()
```

```python
tempt = demographic.loc[demographic['Married']==1]
Married_avg_age = tempt['Age'].mean()
Married_avg_age
```

```python
tempt = demographic.loc[demographic['Married']==0]
UnMarried_avg_age = tempt['Age'].mean()
UnMarried_avg_age
```

```python
#pd.set_option("display.max_rows",None)
demographic['Age'].value_counts().sort_index()
```

```python
for idx,cell in demographic['Married'].iteritems():
    if(np.isnan(cell)):
        if((demographic['Age'][idx]) > (Married_avg_age+UnMarried_avg_age)/2):
            demographic['Married'][idx]=1
        else:
            demographic['Married'][idx]=0
demographic.isnull().sum()
```

```python
for idx,cell in demographic['Dependents'].iteritems():
    if(np.isnan(cell)):
        if((demographic['Married'][idx])):
            demographic['Dependents'][idx]=1
        else:
            demographic['Dependents'][idx]=0
demographic.isnull().sum()
```

```python
tempt = demographic.loc[demographic['Married']==1]
Married_avg_dependents = tempt['Number of Dependents'].mode()[0]
tempt['Number of Dependents'].value_counts()
```

```python
for idx,cell in demographic['Number of Dependents'].iteritems():
    if(np.isnan(cell)):
        if(demographic['Married'][idx]):
            demographic['Number of Dependents'][idx] = Married_avg_dependents
        else:
            demographic['Number of Dependents'][idx] = 0
demographic.isnull().sum()
```
