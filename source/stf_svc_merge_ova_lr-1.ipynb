```python
import pandas as pd
import numpy as np
import random as rand

np.set_printoptions(suppress=True)


def sign(num):
    if (num >= 0):
        return 1
    return -1


def sigmoid(s):
    return 1 / (1 + np.exp((-1) * s))


status_map = {
    'No Churn': 0,
    'Competitor': 1,
    'Dissatisfaction':	2,
    'Attitude': 3,
    'Price': 4,
    'Other': 5
}

statuses = pd.read_csv('../raw/status.csv').to_numpy()
stt_dic = {}
for row in statuses:
    stt_dic[row[0]] = status_map[row[1]]

df = pd.read_csv('stf_svc_merge_train.csv')
labels = []
for idx, row in df.iterrows():
    labels.append(stt_dic[df['Customer ID'][idx]])

df['Churn Category'] = labels
df.drop('Customer ID', axis=1, inplace=True)
dataset = df.to_numpy()

train_size = int(len(dataset) * (4/5))
test_size = len(dataset) - train_size
train_dataset = dataset[:train_size]
test_dataset = dataset[train_size:]

train_dataset_0 = []
train_dataset_1 = []
train_dataset_2 = []
train_dataset_3 = []
train_dataset_4 = []
train_dataset_5 = []
for row in train_dataset:
    train_dataset_0.append([*row[:40], (1 if row[-1] == 0 else 0)])
    train_dataset_1.append([*row[:40], (1 if row[-1] == 1 else 0)])
    train_dataset_2.append([*row[:40], (1 if row[-1] == 2 else 0)])
    train_dataset_3.append([*row[:40], (1 if row[-1] == 3 else 0)])
    train_dataset_4.append([*row[:40], (1 if row[-1] == 4 else 0)])
    train_dataset_5.append([*row[:40], (1 if row[-1] == 5 else 0)])

train_dataset_0 = np.array(train_dataset_0)
train_dataset_1 = np.array(train_dataset_1)
train_dataset_2 = np.array(train_dataset_2)
train_dataset_3 = np.array(train_dataset_3)
train_dataset_4 = np.array(train_dataset_4)
train_dataset_5 = np.array(train_dataset_5)

train_dataset_0_size = len(train_dataset_0)
train_dataset_1_size = len(train_dataset_1)
train_dataset_2_size = len(train_dataset_2)
train_dataset_3_size = len(train_dataset_3)
train_dataset_4_size = len(train_dataset_4)
train_dataset_5_size = len(train_dataset_5)

w_map = {}
for ctg in range(0, 6):
    train_dataset_x = globals()['train_dataset_' + str(ctg)][:, :40]
    train_dataset_y = globals()['train_dataset_' + str(ctg)][:, -1]
    w_t = np.array([0] * 40, dtype=float)
    for _ in range(0, 500):
        sum_gd_e_in = 0
        for i in range(0, globals()['train_dataset_' + str(ctg) + '_size']):
            sum_gd_e_in += sigmoid((-1) * train_dataset_y[i] *
                                   w_t.T.dot(train_dataset_x[i])) * (train_dataset_y[i] * train_dataset_x[i])

        avg_gd_e_in = sum_gd_e_in / train_dataset_0_size
        w_t += 0.1 * avg_gd_e_in

    w_map['w_t_' + str(ctg)] = w_t

test_dataset_x = test_dataset[:, :40]
test_dataset_y = test_dataset[:, -1]
correct_sum = 0
for idx, data in enumerate(test_dataset_x):
    max_score = float('-inf')
    predict = None
    for ctg in range(0, 6):
        score = data.dot(w_map['w_t_' + str(ctg)])
        if score > max_score:
            max_score = score
            predict = ctg
    print(predict, test_dataset_y[idx])
    if predict == test_dataset_y[idx]:
        correct_sum += 1

accuracy = correct_sum / test_size
print(accuracy)
```
